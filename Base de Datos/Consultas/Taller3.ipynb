Ejercicio 2: Crea un nuevo departamento llamado 'AI Research', en la misma location_id que
el departamento con department_id = 90.

```python
import oracledb 
cursor.execute("""
INSERT INTO departments (
    department_id,
    department_name,
    location_id
)
SELECT
    (SELECT MAX(department_id) + 1 FROM departments),
    'AI Research',
    d.location_id
FROM departments d
JOIN locations l
    ON d.location_id = l.location_id
WHERE d.department_id = 9
""")

conn.close()
print("Departamento AI Research agregado correctamente")
```

Ejercicio 3: Agrega un empleado que trabaje en el país 'United States', usando una
subconsulta para obtener una location_id en ese país.

```python
import oracledb 
cursor.execute("""
INSERT INTO employees (
    employee_id,
    first_name,
    last_name,
    email,
    hire_date,
    job_id,
    salary,
    department_id
)
SELECT
    301,
    'John',
    'Smith',
    'JSMITH',
    SYSDATE,
    'IT_PROG',
    7000,
    department_id
FROM departments
WHERE location_id IN (
    SELECT location_id
    FROM locations
    WHERE country_id = (
        SELECT country_id
        FROM countries
        WHERE country_name = 'United States'
    )
)
""")

conn.close()
print("Empleado agregado correctamente")


conn.close()
print("Empleado John Smith agregado correctamente")
```

Ejercicio 4: Agrega un nuevo empleado que tenga el mayor salario posible entre los
empleados con el mismo cargo (job_id = 'IT_PROG').

```python
import oracledb 
cursor.execute("""
INSERT INTO employees (
    employee_id,
    first_name,
    last_name,
    email,
    hire_date,
    job_id,
    salary,
    department_id
)
SELECT
    500,
    'Carlos',
    'Mendoza',
    'CMENDOZA',
    SYSDATE,
    'IT_PROG',
    MAX(salary),
    department_id
FROM employees
WHERE job_id = 16
GROUP BY department_id
FETCH FIRST 1 ROW ONLY""")

conn.close() 
print("Empleado Carlos Mendoza agregado correctamente")
```

```python
UPDATE con condiciones avanzadas
Ejercicio 5: Aumenta el salario de todos los empleados cuyo salario es menor al promedio
de su departamento.
```

```python
import oracledb 
cursor.execute("""
UPDATE employees e
SET e.salary = e.salary * 1.10
WHERE e.salary < (
    SELECT AVG(salary)
    FROM employees
    WHERE department_id = e.department_id
)
""")

conn.close()
print("Salarios actualizados correctamente")
```

Ejercicio 8: Actualiza el job_id de los empleados del departamento 30 al mismo job_id que el
gerente de ese departamento.

```python

import oracledb

cursor = conn.cursor()

cursor.execute("""
UPDATE employees
SET job_id = (
    SELECT job_id
    FROM employees
    WHERE employee_id = (
        SELECT manager_id
        FROM departments
        WHERE department_id = 30
    )
)
WHERE department_id = 30
""")

conn.close()
print("Job_id actualizados correctamente")

```

Ejercicio 11: Elimina a todos los empleados cuyo trabajo sea diferente a 'IT_PROG' y
pertenezcan a un departamento ubicado en 'Europe'.

```python
import oracledb
cursor = conn.cursor()
cursor.execute("""
 DELETE FROM employees
    WHERE job_id NOT IN (
        SELECT job_id FROM jobs
        WHERE job_title = 'Programmer'
    )
    AND department_id in (
        SELECT d.department_id FROM departments d
        JOIN locations l ON d.location_id = l.location_id
        JOIN countries c ON l.country_id = c.country_id
        JOIN regions r ON c.region_id = r.region_id
        WHERE r.region_name = 'Europe'
    )
""")
conn.close()
```
