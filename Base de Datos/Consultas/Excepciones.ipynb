```python
"""
    CODIGO ORIGINAL DE SQL DEVELOPER
    SET SERVEROUTPUT ON
    DEFINE p_partment_desc = 'Information Techology'
    DEFINE p_department_number = 10
    DECLARE
        e_invalid_department EXCEPTION;
    BEGIN
        UPDATE departments
        SET department_name = '&p_department_desc'
        WHERE department_id = &p_department_number;
        IF SQL%NOTFOUND THEN
            RAISE e_invalid_department;
        END IF;
        EXCEPTION 
            WHEN e_invalid_department THEN
                DBMS_OUTPUT.PUT_LINE('No such department id');
    END;
    /
"""
try:
    department_desc = "Information Technology "
    department_number = 300
    cursor.execute(
    f"""
    DECLARE 
        e_invalid_department EXCEPTION;
    BEGIN
        UPDATE          departments
        SET             department_name = '{department_desc}'
        WHERE           department_id = {department_number};
        IF SQL%NOTFOUND then
            RAISE e_invalid_department;
        END IF;
        COMMIT;
    EXCEPTION
        WHEN e_invalid_department THEN
            DBMS_OUTPUT.PUT_LINE('No such department id ');
        WHEN others THEN
            DBMS_OUTPUT.PUT_LINE('Other error type ocured');
    END;
    """)

    line = cursor.var(str)
    status = cursor.var(int)

    while True:
        cursor.callproc("dbms_output.get_line", (line, status))
        if status.getvalue() != 0:
            break
        print(line.getvalue())
except oracledb.Error as e:
    print(f"Ocurrió un error: {e}")
```
