Tipos de variables (globales, locales, hibridas…)
Cuando se llama una variable se tiene de (por valor, por nombre)
PL se puede tener desde cualquier aplicacion

En PL existen 3 tipos de bloques:

SINTAXIS DE DECLARACION DE UNA VARIABLE (todo lo que veas en corchetes es opcional)
SOLO SE PUEDE DECLARAR UNA VARIABLE A LA VEZ

v_hiredate  DATE;
v_deptno    NUMBER(2) NOT NULL :=10;
v_location  CONSTANT NUMBER :=1400; (aqui no es necesario el (4) en el number)

REGLAS EN LOS NOMBRES

no iniciar con numeros ni que haya caracteres especiales

PARA QUE UNA VARIABLE SEA LOCAL EL ORDEN SERIA PRIMERO EL BEGIN Y ADENTRO DE ESE EL DECLARE
SI TIENES UN NUMBER(9,2) te dice que son 9 casillas totales, y que las 2 ultimas estan destinadas a numeros decimales

# %TYPE

v_name      employees.last_name%TYPE; (simplemente ese name tendra el mismo tipo de dato que la columna lastname de employees)

v_numero    v_balance%TYPE := 10;

Cada que quieres crear una variable que se guarde en memoria usas el VARIABLE,

# EJERCICIOS:
## 1
VARIABLE    g_salary    NUMBER;
BEGIN
    SELECT salary
    INTO :g_salary
    FROM employees
    WHERE employee_id = 178;
END;
/
PRINT g_salary;

## 2
SET SERVEROUTPUT ON
DEFINE p_annual_sal = 60000
DECLARE
    v_sal NUMBER(9,2) := &p_annual_sal;
BEGIN
    v_sal := v_sal/12;
    DBMS_OUTPUT.PUT_LINE ('El salario mensual es ' || TO_CHAR(v_sal));
END;
/
