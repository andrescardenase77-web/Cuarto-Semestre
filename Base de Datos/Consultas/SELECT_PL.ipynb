```python
import oracledb
import pandas as pd
import matplotlib.pyplot as plt

try:
    conn = oracledb.connect(
        user = 'hr',
        password = 'basededatos',
        dsn = 'localhost:1521/XE'
    )
except oracledb.Error as e:
    print(e)
    
cursor = conn.cursor()
try:
    cursor.callproc("dbms_output.enable")
    cursor.execute("""
    DECLARE
        v_sum_sal                 employees.salary%TYPE; -- NUMBER(10,2) mejor opcion, si el numero es muy grande
        v_deptno                  employees.department_id%TYPE := 6;
    BEGIN
        SELECT       SUM(salary)
        INTO         v_sum_sal
        FROM         employees
        WHERE        department_id = v_deptno;
        DBMS_OUTPUT.PUT_LINE('Suma de salarios: ' || TO_CHAR(v_sum_sal) || ' numero de departamento: ' || TO_CHAR(v_deptno));
    END;
    """)

    line = cursor.var(str)
    status = cursor.var(int)
    
    while True:
        cursor.callproc("dbms_output.get_line", (line, status))
        if status.getvalue() != 0:
            break
        print(line.getvalue())
except oracledb.Error as e:
    print(f"Error detectado: {e}")
```
