# Ejercicio 3: Agrega un empleado que trabaje en el país 'United States', usando una subconsulta para obtener una location_id en ese país.

```python
import oracledb
conn = oracledb.connect(
    user="HR",
    password="basededatos",
    dsn="localhost:1521/XE"
)
cursor = conn.cursor()

cursor.execute(
    """
    INSERT INTO employees (employee_id, first_name, last_name, email, phone_number, hire_date, job_id, salary, department_id)
    SELECT 304, 'Andres', 'Cardenas', 'ACARDENAS', '12345678', SYSDATE, 19, 3500, department_id FROM departments
    WHERE location_id IN (
        SELECT location_id FROM locations l
        JOIN countries c ON l.country_id = c.country_id
        WHERE country_name = 'United States of America'
    )
    AND rownum = 1
    """
)

print('Se han modificado ',cursor.rowcount,'filas')

cursor.close()
conn.close()
```

# Ejercicio 5: Aumenta el salario de todos los empleados cuyo salario es menor al promedio de su departamento.

```python
import oracledb
conn = oracledb.connect(
    user="HR",
    password="basededatos",
    dsn="localhost:1521/XE"
)
cursor = conn.cursor()

cursor.execute(
    """
    UPDATE employees
    SET salary = salary*1.1
    WHERE salary < (
        SELECT AVG(salary) FROM employees e1
        WHERE employees.department_id = e1.department_id
    )
    """
)

print('Se han modificado ',cursor.rowcount,'filas')

cursor.close()
conn.close()
```

# Ejercicio 7: Cambia el salario de todos los empleados con el puesto 'SA_REP' para que coincida con el salario más bajo que recibe un 'SA_REP'.

# Ejercicio 11: Elimina a todos los empleados cuyo trabajo sea diferente a 'IT_PROG' y pertenezcan a un departamento ubicado en 'Europe'.

```python
import oracledb
conn = oracledb.connect(
    user="HR",
    password="basededatos",
    dsn="localhost:1521/XE"
)
cursor = conn.cursor()

cursor.execute(
    """
    DELETE FROM employees
    WHERE job_id NOT IN (
        SELECT job_id FROM jobs
        WHERE job_title = 'Programmer'
    )
    AND department_id in (
        SELECT d.department_id FROM departments d
        JOIN locations l ON d.location_id = l.location_id
        JOIN countries c ON l.country_id = c.country_id
        JOIN regions r ON c.region_id = r.region_id
        WHERE r.region_name = 'Europe'
    )
    """
)

print('Se han modificado ',cursor.rowcount,'filas')

cursor.close()
conn.close()
```

# Ejercicio 14: Actualiza el salario de todos los empleados que tengan un salario mayor que todos los empleados del departamento 90 para que sea igual al promedio general.

```python
import oracledb
conn = oracledb.connect(
    user="HR",
    password="basededatos",
    dsn="localhost:1521/XE"
)
cursor = conn.cursor()

cursor.execute(
    """
    UPDATE employees
    SET salary  = (
        SELECT AVG(salary) FROM employees
    )
    WHERE salary > (
        SELECT MAX(salary) FROM employees
        WHERE department_id = 9
    )
    """
)

print('Se han modificado ',cursor.rowcount,'filas')

cursor.close()
conn.close()
```
