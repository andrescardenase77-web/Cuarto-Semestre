# 1. Crear un bloque PL/SQL que imprima información sobre el máximo salario pagado a los empleados guarde esta información en una variable y muéstrela en pantalla

```python
try:
    cursor.execute("""
DECLARE
    v_salario_max   employees.salary%TYPE;
BEGIN
    SELECT MAX(salary)
    INTO v_salario_max
    FROM employees;
    DBMS_OUTPUT.PUT_LINE('El salario maximo es: ' || v_salario_max);
END; 
""")

    line = cursor.var(str)
    status = cursor.var(int)
    
    while True:
        cursor.callproc("dbms_output.get_line", (line, status))
        if status.getvalue() != 0:
            break
        print(line.getvalue())
except oracledb.Error as e:
    print(f"Error detectado: {e}")
```

# Crear un bloque PL/SQL para insertar un nuevo departamento en la tabla DEPARTMENTS.
a. Utilice el comando DEFINE para ingresar el código, nombre, del nuevo departamento llamado “EDUCACION”
b. Registre el código location y manager como null por esta vez
c. Muestre en pantalla el nuevo departamento creado

```python
try:
    cursor.execute("""
DEFINE p_codigo = 200;
DEFINE p_nombre = 'EDUCACION';
DECLARE
    v_department_id   departments.department_id%type   := &p_codigo;
    v_department_name departments.department_name%type := '&p_nombre';
BEGIN
    INSERT INTO departments
    (department_id, department_name, location_id)
    VALUES
    (v_department_id, v_department_name, NULL);

    DBMS_OUTPUT.PUT_LINE('Departamento insertado: ' || v_department_id || ' - ' || v_department_name);    
END;
""")

    line = cursor.var(str)
    status = cursor.var(int)
    
    while True:
        cursor.callproc("dbms_output.get_line", (line, status))
        if status.getvalue() != 0:
            break
        print(line.getvalue())
except oracledb.Error as e:
    print(f"Error detectado: {e}")
```
