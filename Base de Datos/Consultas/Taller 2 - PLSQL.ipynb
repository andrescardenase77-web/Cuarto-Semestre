## 3. Realice el siguiente ejercicio
Crear y ejecutar un bloque PL/SQL que acepte dos números, use el comando DEFINE.
Pase los valores al bloque PLSQL y guarde en variables. El primer numero divida por el
segundo numero y adicione el resultado en una variable y muestre en pantalla.

### Resolución

```python
import oracledb

conn = oracledb.connect(
    user="sh",
    password="basededatos",
    dsn="localhost:1521/xe"
)

cursor = conn.cursor()

cursor.callproc("dbms_output.enable")

#Parámetros
num_uno = 10
num_dos = 5

plsql_block = """
DECLARE
    v_num_uno NUMBER := :p_num_uno;
    v_num_dos NUMBER := :p_num_dos;
    v_resultado NUMBER;
BEGIN
    v_resultado := v_num_uno / v_num_dos;
    DBMS_OUTPUT.PUT_LINE ('El resultado es ' || TO_CHAR(v_resultado));
END;
"""

cursor.execute(plsql_block, p_num_uno=num_uno, p_num_dos=num_dos)

lines = []
statusVar = cursor.var(int)
lineVar = cursor.var(str)

while True:
    cursor.callproc("dbms_output.get_line", (lineVar, statusVar))
    if statusVar.getvalue() != 0:
        break
    lines.append(lineVar.getvalue())

for line in lines:
    print(line)

cursor.close()
conn.close()
```

## 4. Realice el siguiente ejercicio
Construir un bloque PLSQL para calcular el total de compensación de un año. El salario
anual y la comisión o bono ganado (porcentaje del salario), se enviara como valor desde el
comando DEFINE p.e.
DEFINE p_salary = 50000
DEFINE p_bomus = 10
….
TOTAL_COMPENSACION → 55000 

### Resolución

```python
import oracledb

conn = oracledb.connect(
    user="sh",
    password="basededatos",
    dsn="localhost:1521/xe"
)

cursor = conn.cursor()


cursor.callproc("dbms_output.enable")

# Parámetros
p_salary = 50000
p_bonus = 10

plsql_block = """
DECLARE
    v_salary   NUMBER := :p_salary;
    v_bonus    NUMBER := :p_bonus;
    v_total    NUMBER;
BEGIN
    v_total := v_salary + (v_salary * v_bonus / 100);

    DBMS_OUTPUT.PUT_LINE('--- RESULTADOS ---');
    DBMS_OUTPUT.PUT_LINE('Salario anual: ' || v_salary);
    DBMS_OUTPUT.PUT_LINE('Porcentaje bono: ' || v_bonus || '%');
    DBMS_OUTPUT.PUT_LINE('Total compensación: ' || v_total);
END;
"""

cursor.execute(plsql_block, p_salary=p_salary, p_bonus=p_bonus)


lines = []
statusVar = cursor.var(int)
lineVar = cursor.var(str)

while True:
    cursor.callproc("dbms_output.get_line", (lineVar, statusVar))
    if statusVar.getvalue() != 0:
        break
    lines.append(lineVar.getvalue())

for line in lines:
    print(line)

cursor.close()
conn.close()
```
