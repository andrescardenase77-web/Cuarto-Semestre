```python
import oracledb

conn = oracledb.connect(
    user="HR",
    password="basededatos",
    dsn="localhost:1521/XE"
)

cursor= conn.cursor()
cursor.execute("SELECT sum(salary), department_id FROM employees group by department_id order by department_id")

resultados= cursor.fetchall()

salarios=[]
departamentos=[]

"""cada fila de ese resultado tiene dos columnas: la suma de salarios y el id del departamento"""
for fila in resultados:
    salarios.append(fila[0])
    departamentos.append(fila[1])
import matplotlib.pyplot as pit

pit.figure(figsize=(10,5))
bars= pit.bar(departamentos, salarios)

pit.title("Salarios totales pagados por departamento")
pit.xlabel("ID del Departamento")
pit.ylabel("Salario Total ($)")

# agregar etiquetas de valor encima de cada barra

for i, bar in enumerate(bars):
    altura= bar.get_height()
    pit.text(bar.get_x() + bar.get_width()/2,
    altura + 1000, 
    f"Depart({departamentos[i]})",
    ha='center', va='bottom', fontsize=9)

pit.tight_layout()
pit.show()
```
