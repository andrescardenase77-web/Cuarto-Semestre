# 3. Realizar el siguiente bloque plsql: declare dos variables una tipo character y otra tipo numerica,
# inicialícelas, envie a imprimir por pantalla los valores asignados a las variables

```python
import oracledb

conn = oracledb.connect(
    user="hr",
    password="basededatos",
    dsn="localhost:1521/xe"
)

cursor = conn.cursor()


cursor.callproc("dbms_output.enable")


cursor.execute("""
DECLARE
    v_texto   VARCHAR2(30) := 'Hola mundo';
    v_numero  NUMBER := 25;
BEGIN
    DBMS_OUTPUT.PUT_LINE('Valor de la variable character: ' || v_texto);
    DBMS_OUTPUT.PUT_LINE('Valor de la variable numerica: ' || v_numero);
END;
""")


status = cursor.var(oracledb.NUMBER)
line = cursor.var(oracledb.STRING)

while True:
    cursor.callproc("dbms_output.get_line", [line, status])
    if status.getvalue() != 0: 
        break
    print(line.getvalue())

cursor.close()
conn.close()
```
